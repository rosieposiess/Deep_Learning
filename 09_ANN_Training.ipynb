```python
import numpy as np
import matplotlib.pyplot as plt
```

## Data for Linear Regression

```python
x = np.arange(0, 10, 0.5)
t = np.array([-1.52129006, -1.27491772, -0.96940479, -0.40476183,  0.98989321,
              0.34758961,  1.39646606,  1.50900255,  2.59080334,  2.77800334,
              3.45137288,  4.7229806 ,  4.7882944 ,  4.96923468,  5.73584226,
              6.1385543 ,  7.17248654,  8.29641142,  7.59675138,  8.72590778])
```

```python
plt.plot(x, t, '*')
plt.show()
```

```python
x.shape #총 데이터
```

## One Layer Network Training

```python
# the number of data
N = x.shape[0]


# initial weight and bias
w = np.random.randn(1)
b = np.random.randn(1)


# learning rate
lr = 0.03

loss_list = []

# training
for i in range(500):
    pred_y = w*x + b   # forward propagation
    loss = np.sum((pred_y - t)**2)/N   # MSE loss
    loss_list.append(loss)

    print('%3d iteration,  loss=%.5f' %(i+1, loss))

    # 미분
    grad_w = (2/N) * np.dot(x, (pred_y - t))  # analytic gradients
    grad_b = (2/N) * np.sum(pred_y - t)

    # 경사하강법
    w -= lr*grad_w
    b -= lr*grad_b

print('='*30)
print('linear regression:  y = %.3f x + %.3f' %(w, b))
```

```python
plt.plot(loss_list)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()
```

## Result

```python
plt.plot(x, t, '*')

xx = np.arange(0, 10, 0.1)
plt.plot(xx, w*xx+b)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(x.reshape(-1, 1), t.reshape(-1, 1))
print(reg.coef_, reg.intercept_)
```
