```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2
```

## Convolution

```python
# kernel(filter)
kernel1 = np.ones((3, 3))/9
kernel2 = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
kernel3 = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
print(kernel1)
print(kernel2)
print(kernel3)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img = imread('/content/drive/MyDrive/Colab Notebooks/cactus.png')
```

```python
img.shape
```

```python
# cv2.filter2D(입력영상, 출력영상 데이터 타입, 커널)
img1 = cv2.filter2D(img, -1, kernel1)
img2 = cv2.filter2D(img, -1, kernel2)
img3 = cv2.filter2D(img[:,:,:3], -1, kernel3)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('original')
plt.subplot(1, 2, 2)
plt.imshow(img1)
plt.title('kernel 1')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('original')
plt.subplot(1, 2, 2)
plt.imshow(img2)
plt.title('kernel 2')
plt.show()
```
