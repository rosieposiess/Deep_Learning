```python
# reference
a = [1.0, 2.0, 3.0]
b = a               # b is a reference.
b[0] = 5.0          # Change 'b'
print('a =', a, '\nb =', b)            # The change is reflected in 'a'
```

```python
# independent copy
c = a[:]          # c = a.copy(), 'c' is an independent copy of 'a'
c[0] = 1.0        # Change 'c'
print('a =', a, '\nc =', c)          # 'a' is not affected by the change
```

## 3. 제어문 (control)

- 조건문: if-else
- 반복문: for

### 3.1 if-else 문

```python
my_score = 78

if my_score >= 90:
    my_grade = 'A'
elif my_score >= 80:
    my_grade = 'B'
elif my_score >= 70:
    my_grade = 'C'
elif my_score >= 60:
    my_grade = 'E'
else:
    my_grade = 'F'

print(my_grade)
```

```python
# one line 조건문
my_grade_state = 'Pass' if my_score >= 60 else 'Fail'
print(my_grade_state)
```

### 3.2 for 문

```python
# 반복문 - for
for i in [0, 1, 2, 3, 4]:
    print(i)
```

```python
for i in range(5):
    print(i)
```

```python
S = 0
for j in range(1, 11, 2):
    print(j)
    S = S + j
```

```python
print(S)
```

```python
fruits = ['apple', 'banana', 'cherry']

for i in fruits:
    print(i)
```

```python
for index in range(len(fruits)):
    print(index, fruits[index])
```

```python
for index, value in enumerate(fruits):
    print(index, value)
```

```python
nums = [0, 1, 2, 3, 4]
squares = []
for x in nums:
    squares.append(x ** 2)  # squares list에 x**2 원소 추가
print(squares)
```

```python
# 확장 문법 - 반복문
nums = [0, 1, 2, 3, 4]
squares = [x ** 2 for x in nums]
print(squares)
```

```python
# 확장 문법 - 반복문 + 조건문
nums = [0, 1, 2, 3, 4]
even_squares = [x ** 2 for x in nums if x % 2 == 0] # equal =, not equal !=
print(even_squares)
```

```python
# in a dictionary
d = {'person': 2, 'cat': 4, 'spider': 8}
for animal, legs in d.items():
    print('A %s has %d legs' % (animal, legs))
```

```python
nums = [0, 1, 2, 3, 4]
even_num_to_square = {x: x ** 2 for x in nums if x % 2 == 0}
print(even_num_to_square)
```

### 3.3 while 문

```python
fn_list = [1, 1]
while len(fn_list) < 20:
    fn_list.append(fn_list[-1] + fn_list[-2])
print(len(fn_list))
```
