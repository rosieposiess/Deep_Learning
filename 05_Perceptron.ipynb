```python
import numpy as np
```

## 1. AND 연산

```python
def AND(x1, x2):
    X = np.array([1, x1, x2])
    W = np.array([-2, 1.5, 1])
    if np.dot(X, W) <= 0:   # np.dot(X, W.T) = np.sum(X*W)
        return 0
    else:
        return 1
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(x[0], x[1])
    print("AND", x, " = ", y)
```

## 2. Not AND 연산

```python
def NAND(x1, x2):
    X = np.array([1, x1, x2])
    W = np.array([2, -1.5, -1])

    if np.dot(X, W) <= 0:   # np.dot(X, W.T) = np.sum(X*W)
        return 0
    else:
        return 1
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = NAND(x[0], x[1])
    print("NAND", x, " = ", y)
```

## 3. OR 연산

```python
def OR(x1, x2):
    X = np.array([1, x1, x2])
    W = np.array([-0.5, 1, 1])

    if np.dot(X, W) <= 0:   # np.dot(X, W.T) = np.sum(X*W)
        return 0
    else:
        return 1
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = OR(x[0], x[1])
    print("OR", x, " = ", y)
```

## 4. XOR 연산

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)

    return AND(s1, s2)
```

```python
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("XOR", x, " = ", y)
```
