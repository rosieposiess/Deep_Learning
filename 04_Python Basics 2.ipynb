### 3.3 함수 생성

```python
#1부터 n까지 합을 구하는 함수를 작성하기
# 함수 def 구조
# def ftn_name(var1, var2, ...):
# ....statement1
# ....statement2
# ....(return something)

def sum_to_n(n):
    S = 0
    for j in range(1, n+1):
        S += j
    return S
```

```python
a = sum_to_n(10)
print(a)
```

```python
# return 의 역할

def sum_to_n2(n):
    S = 0
    for j in range(1, n+1):
        S += j
    print(S)

b = sum_to_n2(10)
```

## 4. Class

붕어빵틀 -> 붕어빵

Class ->  인스턴스 생성

```python
class Shape:
    # constructor (생성자)
    def __init__(self, width, height):
        self.width = float(width)  # Create an instance variable
        self.height = float(height)
        print('Initialized!')

    # method
    def show(self):
        print('width: %.2f, height: %.2f 도형입니다.' % (self.width, self.height))

    def add_width(self, width):
        self.width += width
```

```python
# 인스턴스(instance) 생성
g = Shape(10, 20)
# init, 즉 생성자가 실행된다.
```

```python
g.show()
```

```python
g.add_width(7)
g.show()
```

```python
# 상속

class Triangle(Shape):
    # 메쏘드
    def area(self):
        return (self.width * self.height) / 2

class Square(Shape):
    # 메쏘드
    def area(self):
        return self.width * self.height
```

```python
t = Triangle(10, 20)  # 인스턴스 생성
```

```python
t.show()
```

```python
print(t.area())
```

```python
s = Square(10, 20)  # 인스턴스 생성
s.add_width(10)
```

```python
s.show()
```

```python
print(s.area())
```
